1. Define un método `sum(array)` que tome un arreglo de números enteros como argumento y devuelva la suma de sus elementos. Para una matriz vacía, debería devolver cero. 

Solucion: Si bien es cierto esto se podria realizar con un bucle para solucionar este problema, vamos a usar el metodo `inject`, un metodo de numeración como sigue:

```ruby
def suma arr
  arr.inject(0) {|acomulador, numero| acomulador + numero}
end
```

```ruby
suma([1,2,3])
```

Se puede usar una forma mas consisa para realizar el programa anterior, de la siguiente manera:

```ruby
def suma arr
  arr.inject(0, :+)
end
```

```ruby
suma([1,2,3,4,5])
```

Aqui solo utilizamos como segundo argumento el simbolo de la suma, haciendo referencia que queremos sumar los elementos del array comenzando dede el 0 (eso quiere decir que si arr es vacio entonces por defecto retorna 0)

```ruby
suma([])
```

2. Define un método max_2_sum(array) que tome un arreglo de números enteros como argumento y devuelva la suma de sus dos elementos más grandes. Para un arreglo vacío, debería devolver cero. Para un arreglo con solo un elemento, debería devolver ese elemento (considera si los dos elementos más grandes también tienen el mismo valor). 

Solución: Para solucionar este problema vamos a usar dos condicionales para satisfacer si el arreglo es vacío o si tiene un solo elemento. Luego vamos a ordenar el arreglo con sort y por ultimo sumamos los dos ultimos elementos

```ruby
def max_2_sum arr
  return 0 if arr.empty?
  return arr[0] if arr.size == 1
  arr = arr.sort
  arr[-1] + arr[-2]
end
```

```ruby
puts max_2_sum([])
puts max_2_sum([5])
puts max_2_sum([1,2,3,4,5])
puts max_2_sum([3,7,2,6,9,1])
puts max_2_sum([4,1,4,8,5,3,6,8,5])
```

3. Define un método sum_to_n?(array, n) que toma un arreglo de números enteros y un número entero adicional n, como argumentos y devuelva verdadero si dos elementos cualesquiera en el arreglo de enteros suman n. sum_to_n?([], n) debería devolver false para cualquier valor de n, por definición.

En este ejercicio podemos usar el metodo combination junto a any? para buscar algunos valores tal que cumplan la condicion:

```ruby
def sum_to_n? arr, n
  return false if arr.empty?
  res = arr.combination(2).any?{|x,y| n == x+y}
  res
end
```

```ruby
puts sum_to_n?([],3)
puts sum_to_n?([1,2,3,4,5,6,7,8],8)
puts sum_to_n?([2,4,5,5,1],10)
puts sum_to_n?([1,2,8,3],100)
```

2. Define un método start_with_consonant?(s) que tome una cadena y devuelva verdadero si comienza con una consonante y falso en caso contrario. Asegúrate de que funcione tanto para mayúsculas como para minúsculas y para otras letras. 

Realizamos tres comprobaciones, primero verificamos si la cadena esta vacia, esto deberia retornar false ya que no empieza con una consonante, luego verificamos si la primera letra de la cadena es una letra, si no lo es retornara falso. Por ultimo verificamos si empieza con una vocal sin importar si se encuentra en mayusculas o no, si verifica retorna falso sino sera verdadero

```ruby
def starts_with_consonant? s
  return false if s.empty?
  letra = /^[\w]/
  return false if !s.match(letra)
  vocales = /^[aeiou]/i
  if s.match(vocales) then false else true end
end
```

```ruby
puts starts_with_consonant?("")
puts starts_with_consonant?("Hola")
puts starts_with_consonant?("Amelia")
puts starts_with_consonant?("roberto")
puts starts_with_consonant?("orlando")
puts starts_with_consonant?("%otro")
```

3. Define un método binary_multiple_de_4?(s) que tome una cadena y devuelva verdadero si la cadena representa un número binario que es múltiplo de 4, como '1000'. Asegúrate de que devuelva falso si la cadena no es un número binario válido. Ejecuta las pruebas asociadas a través de:

Solucion: Primero verificamos si el String tiene la forma de un numero binario, luego lo convertimos a un numero entero de base 10 y verificamos si es divisible por 4, si lo es retornara true sino false

```ruby
def binary_multiple_of_4? s
  patron = /^[01]+$/
  if patron.match(s) then s.to_i(2)%4==0 else false end
end
```

```ruby
puts binary_multiple_of_4?("")
puts binary_multiple_of_4?("10")
puts binary_multiple_of_4?("100")
puts binary_multiple_of_4?("101")
puts binary_multiple_of_4?("1111")
puts binary_multiple_of_4?("1010")
puts binary_multiple_of_4?("10100")
```

Define una clase BookInStock que representa un libro con un número ISBN, isbn y el precio del libro como número de punto flotante, precio, como atributos.

```ruby
class BookInStock
  attr_accessor :isbn, :price
  def initialize(isbn, price)
    @isbn = isbn
    @price = price
  end
end
```

Verificamos:

El constructor debe aceptar el número ISBN (una cadena, ya que en la vida real los números ISBN pueden comenzar con cero e incluir guiones) como primer argumento y el precio como segundo argumento y debe generar ArgumentError (una de los tipos de excepciones integradas de Ruby) si el número ISBN es la cadena vacía o si el precio es menor o igual a cero. Incluye los getters y setters adecuados para estos atributos. 

```ruby
class BookInStock
  attr_accessor :isbn, :price
  def initialize(isbn, price)
    raise ArgumentError.new("Numero ISBN no valido") if isbn.length ==0
    raise ArgumentError.new("El precio no puede ser 0") if price <= 0
    @isbn = isbn
    @price = price
  end
end
```

Verificamos:

Incluye un método price_as_string que devuelva el precio del libro formateado con un signo de dólar inicial y dos decimales, es decir, un precio de 20 debe tener el formato `$20.00` y un precio de 33.8 debe tener el formato `$33.80`. Consulta los métodos de cadenas formateadas en Ruby.

```ruby
class BookInStock
  attr_accessor :isbn, :price
  def initialize(isbn, price)
    raise ArgumentError.new("Numero ISBN no valido") if isbn.length ==0
    raise ArgumentError.new("El precio no puede ser 0") if price <= 0
    @isbn = isbn
    @price = price
  end
  def price_as_string
    "$%.2f" % @price
  end
end
```
